### 1. Read MNIST using Keras

```python
from utils import load_mnist
(x_train, y_train), (x_test, y_test) = load_mnist()
```

## 1.  LetNet AdaDelta Relu

```python
from keras.utils.vis_utils import plot_model # can be used to plot the model into a png file.
from LeNet import LeNet
NUM_CLASSES = 10
model = LeNet(num_classes = NUM_CLASSES)
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
# train model
BATCH_SIZE = 128
MAX_EPOCH = 12

hist = model.fit(x_train, y_train,
          batch_size=BATCH_SIZE,
          epochs=MAX_EPOCH,
          verbose=0,
          validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])

from matplotlib import pyplot as plt
fig = plt.figure(figsize =(11,5))
plt.subplot(121)
plt.plot([i for i in range(MAX_EPOCH)], hist.history["loss"], label="Training Loss", color="orange")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.subplot(122)
plt.plot([i for i in range(MAX_EPOCH)], hist.history["val_loss"], label="Test Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show();
```
